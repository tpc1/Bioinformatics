```python
t = gene.count('T')
print('Number of thyrosine bases '+str(t))
```

```python
g = gene.count('G')
print('Number of guanine bases '+str(g))
```

```python
start = gene.count('ATG')
print('Nunber of start codons is',+start)
```

```python
stop = 0
for i in ('TAG','TAA','TGA'):
  stop += gene.count(i)
print('Nunber of stop codons is',+stop)
```
